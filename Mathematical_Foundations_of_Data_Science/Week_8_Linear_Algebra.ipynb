### What Is Matrix Transposition?
#### Definition: matrix transposition

Given a matrix **A** with dimensions $m \times n$, its transposition $A^T$ is a matrix with dimensions $n \times m$ such that:

$$
(A^T)_{ij} = A_{j,i}
$$

for example

$$
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}^T
=
\begin{bmatrix}
1 & 4 \\
2 & 5 \\
3 & 6
\end{bmatrix}
$$

Basically, transposing a matrix flips it along the diagonal starting from the top left (0, 0) corner. The transposed matrix does not need to be square; the lengths along each axis may be different.

#### Example: Square Transposition
$$
\begin{bmatrix}

\end{bmatrix}

### What is Matrix Multiplication?
Matrix multiplication is one of the most common operations in linear algebra. In the uses below, you will see that artificial intelligence is pushing matrix multiplication to be one of the top uses of compute power. Direct use cases for matrix multiplication and its central position in linear algebra will be covered in later weeks.

#### Definition: Matrix Multiplication
Given matrix **A** with dimensions $m \times n$ and matrix **B** with dimensions $n \times o$, then their matrix product **AB** is a matrix with dimensions $m \times o$ such that

$$
(AB)_{i,j} = \sum_{k=0}^{n-1} A_{i,k} B_{k,j}
$$

The process of computing the matrix product is called matrix multiplication.

A direct implementation of matrix multiplication takes time proportional to $mno$, but faster methods are possible for large matrices.

### What is an Identity Matrix?
An identity matrix is a matrix $I$ that acts as an identity element with matrix multiplication.

#### Definition: Identity Matrix
An identity matrix is an matrix $n \times n$ matrix $I_n$ such that for any $m \times n$ matrix **A** or any $n \times m$ matrix **B**.

$$
AI_n = A
$$
$$
I_{n}B = B
$$

Matrix multiplication with an identity matrix always returns the other matrix.

It is easy to confirm that an identity matrix must be square, have all ones on the diagonal, and be zero elsewhere.

#### Example: 3x3 Identity Matrix
$$
I_3 = \begin{bmatrix}
1 & 0 & 0\\
0 & 1 & 0\\
0 & 0 & 1
\end{bmatrix}
$$

For convenience, the variable $I$ is often used to refer to all identity matrices regardless of size, and the specific identity matrix is determined by context from the dimensions.

### What is Matrix Inversion?
#### Definition: Matrix Inverse (square only)

For $n \times n$ matrix $A$, the matrix inverse of $A$, denoted $A^-1$, is an $n \times m$ matrix such that
$$
AA^-1 = I_n
$$
$$
A^{-1}A = I_n
$$

Matrix inverses do not always exist. Matrix inverses may exist for non-square matrices but have somewhat different behavior. It is possible to algorithmically determine whether a matrix inverse exists and to compute it. More detail about these will be covered in later weeks.

Linear algebra starts with the study of linear equations. Systems of linear equations are straightforward to solve but have many practical applications. Systems of linear equations also are components of more general equation solvers and optimization methods.

### Systems of Linear Equations
What are linear equations?

A linear equation with $n$ variables $X_1,\dots,X_n$ has the form:

$$
\sum^{n-1}_{j=0}a_jx_j = b
$$

Using vectors $a$ and $x$ to represent the coefficients and variables respectively. 

This can be expressed with a matrix multiplication as:

$$
a^Tx = b
$$

The key elements are that each variable can be multiplied by its own coefficient, but not other variables.

Here, the variables are the $x_j$ and the coefficient of $x_j$ is a $_j$.

The whole equation says that the sum of the products of variables and their coefficients, that is the sum of all the $a_j$'s and $x_j$'s = $b$.

For two dimensions, it is common to use:
$$
ax + by = c
$$

or, one coefficient for the x variable in constant b:

$$
y = mx + b
$$

The first equation is equivalent for the general equation. The second version ($y=mx+b$) is a common equation for a line, but does not work for vertical lines.

## ax + by = c
- if $a \neq 0$, the $x$ intercept is $\frac{c}{a}$ when $y = 0$
- if $b \neq 0$, the $x$ intercept is $\frac{c}{b}$ when $x = 0$
- the cases $a=0$ and $b=0$ correspond to horizontal and vertical lines respectively.

figuring out the x and y intercepts is easy in this form.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100)  # Smoother line with more points
y = 3 - (2 * x)

plt.plot(x, y, label="2x + y = 3")
plt.plot(x, [0 for _ in x], color='black')  # X-axis
plt.axvline(x=0, color='black')  # Y-axis

# Adjusted label positions using plt.annotate
plt.annotate('y-intercept\n(0, 3/1)', xy=(0, 3), xytext=(-40, 10),
             textcoords='offset points', fontsize=12, ha='right', arrowprops=dict(arrowstyle="->"))
plt.annotate('x-intercept\n(3/2, 0)', xy=(1.5, 0), xytext=(10, -30),
             textcoords='offset points', fontsize=12, ha='left', arrowprops=dict(arrowstyle="->"))

# Add the legend
plt.legend(loc='upper right', fontsize=12, fancybox=True, framealpha=0.5)

plt.xlim([-2, 2])  # Set X-axis limits
plt.ylim([-1, 7])  # Set Y-axis limits to match your image style
#plt.grid(True)  # Optional grid if you prefer
plt.show()
```

A system of linear equations is a set of linear equations:
- The system of linear equations represents all the variable assignments that satsify all the equations at once.
- We call finding those assignments "solving the system of linear equations".

We can describe systems of lienar equations like this:
$$\forall_i \sum_{j=0}^{n-1} a_{i,j} x_j = b_i$$
For all i, the sum from $j=0$ to $n-1$ of $a_i$ $_jX_j$ is $b_i$

With matrix multiplication, we can write that as
$$
Ax = b
$$
with $A$ being a matrix of $a_i,_j$ values, $x$ a vector of the variables, and $b$ a vector of the constants. To describe a system of linear equations, we are essentially repeating the equation for one linear equation, but adding some extra subscripts to distinguish them.

Of course, for a particular set of equations, we will write out all the individual equations with the sum unrolled. Let's look at an example with lines again. 

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbols for x and y
x, y = sp.symbols('x y')

# Define the two linear equations
eq1 = sp.Eq(2 * x + y, 3)  # 2x + y = 3
eq2 = sp.Eq(-x + 4 * y, 2)  # -x + 4y = 2

# Solve the system of equations
solution = sp.solve([eq1, eq2], (x, y))
solution_x = solution[x]
solution_y = solution[y]

# Create a range of x values for plotting
x_vals = np.linspace(-2, 2, 100)

# Convert sympy expressions to lambda functions for numerical evaluation
eq1_y_vals = sp.lambdify(x, sp.solve(eq1, y)[0], 'numpy')(x_vals)
eq2_y_vals = sp.lambdify(x, sp.solve(eq2, y)[0], 'numpy')(x_vals)

# Plot the equations
plt.plot(x_vals, eq1_y_vals, label=r'$2x + y = 3$')
plt.plot(x_vals, eq2_y_vals, label=r'$-x + 4y = 2$')

# Mark the solution (intersection point)
plt.scatter(float(solution_x), float(solution_y), color='black', zorder=5, label='solution', marker='*', s=100)
plt.axvline(x=0, color='black',linewidth=0.5)  # Y-axis
plt.axhline(y=0, color='black',linewidth=0.5)  # X-axis

# Add annotations
plt.annotate(f"({solution_x:.2f}, {solution_y:.2f})", 
             (float(solution_x), float(solution_y)), 
             textcoords="offset points", xytext=(10,10), ha='center')

# Add labels, grid, and legend
plt.xlim([-2, 2])
plt.ylim([-1, 7])
plt.legend(loc='upper right')
#plt.grid(True)
plt.show()
```

you can see each equation graphically represented as a line, and the solution to the system of linear equations is the point at the intersection.

#### Standard Form of Linear Programs
Definition: Linear Programming (standard form)
Given:
$$A: m \times n$$
$$b : m \times 1$$
$$ c: n \times 1$$
Constraints:
$$ Ax \leq b$$
$$ 0 \leq x$$
Find $x$ : $n \times 1$ maximizing $c^Tx$

#### Augmented Form of Linear Programs
Definition: Linear Programming (Augmented form)
Given:
$$A: m \times n$$
$$b : m \times 1$$
$$ c: n \times 1$$
Constraints:
$$ Ax = b$$
$$ 0 \leq x$$
Find $x$ : $n \times 1$ maximizing $c^Tx$

#### Quadratic Programming 
Definition: Quadratic Programming (Augmented form)
Given:
$$A: m \times n$$
$$b : m \times 1$$
$$ c: n \times 1$$
$$ \sum : n \times n$$
Constraints:
$$ Ax \leq b$$
$$ 0 \leq x$$
Find $x$ : $n \times 1$ minimizing $c^Tx + \frac{1}{2} X^T \sum X$

The quadratic programming funciton is the same as the linear programming function, but another matrix sigma is added to make the objective function quadratic, and the optimization is usually minimizing, not maximizing.

To solve the finanical risk problem, the c vector is the expected returns of each asset, and the sigma matrix is the covariance matrix of those assets.

To make a function "quadratic" means to write or modify the function such that it follows the standard form of a quadratic function, which is a polynomial of degree 2. A quadratic function typically takes the form:

$$
f(x) = ax^2 + bx + c
$$

where:
- \( a \), \( b \), and \( c \) are constants (with \( a \neq 0 \)),
- \( x \) is the variable,
- \( ax^2 \) is the quadratic term (which gives the function its degree 2),
- \( bx \) is the linear term, and
- \( c \) is the constant term (also called the intercept).

### Key Characteristics of a Quadratic Function:
- **Degree of 2**: The highest exponent of \( x \) is 2.
- **Parabolic shape**: The graph of a quadratic function is a parabola, which can either open upwards (if \( a > 0 \)) or downwards (if \( a < 0 \)).
- **Vertex**: The parabola has a vertex, which is either the minimum or maximum point depending on the sign of \( a \).
- **Axis of symmetry**: The parabola is symmetric about a vertical line passing through the vertex.

### Example:
The function \( f(x) = 2x^2 + 3x + 1 \) is quadratic because:
- The degree of the function is 2 (due to the \( 2x^2 \) term).
- Its graph is a parabola that opens upwards (since \( a = 2 > 0 \)).

### How to Make a Function Quadratic:
1. **Add a quadratic term**: If you have a linear function, such as \( f(x) = bx + c \), you can make it quadratic by adding a \( ax^2 \) term, where \( a \neq 0 \).
   - For example, if \( f(x) = 3x + 5 \), adding a quadratic term results in \( f(x) = 2x^2 + 3x + 5 \).
   
2. **Ensure the degree is 2**: A function becomes quadratic when its highest degree (power of \( x \)) is 2. Functions of higher degree (e.g., cubic functions) or lower degree (e.g., linear functions) are not considered quadratic.

### Applications:
Quadratic functions are common in physics (projectile motion), economics (profit maximization), and engineering (optimization problems), as they describe phenomena that involve squared terms, such as acceleration and areas.

## Linear Model Recap
A linear model has the basic form

$f(\mathbf{x}) = \sum_{j=0}^{n-1} (c_j * x_j) + c_n$

or using matrix multiplication and appending constant 1 to $\mathbf{x}$,

$f(\mathbf{x}) = \mathbf{c}^T \mathbf{x}$

Given training input $\mathbf{X}$ with corresponding outputs $\mathbf{Y}$, the ordinary least squares linear regression is

$f(\mathbf{x}) = \beta \mathbf{x}$

$\beta = ( \mathbf{X}^T \mathbf{X} )^{-1} \mathbf{X}^T \mathbf{Y}$

While we still have not justified that formula, you can at least observe that it is composed of the linear algebra operations just introduced.

### Building More Complex Models With Linear Model Solvers
This model demonstrates how to build more complex models using code for linear solvers.

Linear models are pretty rigid in their form... one coefficient per variable plus an additive constant, no more, no less. 

So if you want to regress a different kind of model, say a quadratic model or any polynomial, then those functions cannot be expressed with the functional form of a linear model.

#### Reusing Linear Model Solvers
The linear models that we have built look like:
$$f(x) = c_0x_0 + c_1x_1 + \dots + c_n$$

But... there is a trick. If we change the variables going into the linear model solver, then we can have the linear model solver regress the best coefficients. This is basically designing new variables based on the more complex functions, so we solve a linear model taking in polynomial variables.

#### Motivation for Principal Components
This video briefly motivates principal components.

#### Principal Component Analysis:
- Identifies the directions of highest variance in a dataset
- Builds a new coordinate system based on those directions

Part of how this works is that it just looks at the numbers, no labels, no units, no semantics at all. Then we imagine those coordinates to be positions in space and try to figure out the direction which will maximize the variance.

#### Definining Principal Components: First Principal Component
Given a dataset $X$, then the first principal component of $X$, denoted $w_{(1)}$, is chosen so that
$$
w_{(1)} = argmax_{\|\mathbf{w}\| =1}var(Xw)
$$
That is, $w_{(1)}$ is the unit vector maximizing the variance of the vector $Xw$.

Each principal component is a unit vector that is orthogonal to the previous principal components and maximizes the variance in the data that is not explained by the previous principal components.

We will develop the necessary machinery to compute that vector over the course of this module.

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Step 1: Create a simple dataset
np.random.seed(0)
data = np.random.rand(100, 2)  # 100 samples, 2 features
df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])

# Step 2: Perform PCA on the original dataset
pca = PCA(n_components=2)
pca_original = pca.fit_transform(df)

# Store the original principal components
original_pc1 = pca.components_[0]
original_pc2 = pca.components_[1]

# Step 3: Add a constant (100) to the first column and perform PCA
df_add_constant = df.copy()
df_add_constant['Feature1'] += 100
pca_add_constant = PCA(n_components=2).fit(df_add_constant)

# Step 4: Multiply the first column by 5 and perform PCA
df_multiply_scalar = df.copy()
df_multiply_scalar['Feature1'] *= 5
pca_multiply_scalar = PCA(n_components=2).fit(df_multiply_scalar)

# Plot the principal component directions
plt.figure(figsize=(8, 6))

# Original principal components
plt.quiver(0, 0, original_pc1[0], original_pc1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original PC1')
plt.quiver(0, 0, original_pc2[0], original_pc2[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='Original PC2')

# Principal components after adding constant
plt.quiver(0, 0, pca_add_constant.components_[0][0], pca_add_constant.components_[0][1], angles='xy', scale_units='xy', scale=1, color='green', label='Add Constant PC1')
plt.quiver(0, 0, pca_add_constant.components_[1][0], pca_add_constant.components_[1][1], angles='xy', scale_units='xy', scale=1, color='lightgreen', label='Add Constant PC2')

# Principal components after multiplying by scalar
plt.quiver(0, 0, pca_multiply_scalar.components_[0][0], pca_multiply_scalar.components_[0][1], angles='xy', scale_units='xy', scale=1, color='red', label='Multiply by 5 PC1')
plt.quiver(0, 0, pca_multiply_scalar.components_[1][0], pca_multiply_scalar.components_[1][1], angles='xy', scale_units='xy', scale=1, color='orange', label='Multiply by 5 PC2')

# Plot settings
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title("Effect of Constant Addition vs. Multiplication on Principal Components")
plt.grid(True)
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbols for x and y
x, y = sp.symbols('x y')

# Define the two linear equations
eq1 = sp.Eq(2 * x + y, 3)  # 2x + y = 3
eq2 = sp.Eq(-x + 4 * y, 2)  # -x + 4y = 2

# Solve the system of equations
solution = sp.solve([eq1, eq2], (x, y))
solution_x = solution[x]
solution_y = solution[y]

# Create a range of x values for plotting
x_vals = np.linspace(-2, 2, 100)

# Convert sympy expressions to lambda functions for numerical evaluation
eq1_y_vals = sp.lambdify(x, sp.solve(eq1, y)[0], 'numpy')(x_vals)
eq2_y_vals = sp.lambdify(x, sp.solve(eq2, y)[0], 'numpy')(x_vals)

# Plot the equations
plt.plot(x_vals, eq1_y_vals, label=r'$2x + y = 3$')
plt.plot(x_vals, eq2_y_vals, label=r'$-x + 4y = 2$')

# Mark the solution (intersection point)
plt.scatter(float(solution_x), float(solution_y), color='black', zorder=5, label='solution', marker='*', s=100)
plt.axvline(x=0, color='black',linewidth=0.5)  # Y-axis
plt.axhline(y=0, color='black',linewidth=0.5)  # X-axis

# Add annotations
plt.annotate(f"({solution_x:.2f}, {solution_y:.2f})", 
             (float(solution_x), float(solution_y)), 
             textcoords="offset points", xytext=(10,10), ha='center')

# Add labels, grid, and legend
plt.xlim([-2, 2])
plt.ylim([-1, 7])
plt.legend(loc='upper right')
#plt.grid(True)
plt.show()
```
